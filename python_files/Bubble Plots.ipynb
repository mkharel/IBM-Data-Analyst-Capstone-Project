### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here
age_mapping = {
                "25-34 years old": 29.5,
                "35-44 years old": 39.5,
                "18-24 years old": 19.5,
                "45-54 years old": 49.5,
                "55-64 years old": 59.5,  
                "Under 18 years old": 17,
                "65 years or older": 65,
                "Prefer not to say": 0  
             }
SOPartFreq_mapping = {
                        "Less than once per month or monthly": 1,
                        "I have never participated in Q&A on Stack Overflow": 2,
                        "A few times per month or weekly": 3,
                        "A few times per week": 4,
                        "Daily or almost daily": 5,
                        "Multiple times per day": 6
                    }
df['Age_numeric'] = df['Age'].map(age_mapping)
df['SOPartFreq_numeric'] = df['SOPartFreq'].map(SOPartFreq_mapping)
#df = df.dropna(subset=['Age', 'SOPartFreq', 'JobSat'])

plt.figure(figsize=(8,6))
plt.scatter(df['Age_numeric'], df['SOPartFreq_numeric'], 
            s=df['JobSat']*5,
            alpha=0.6,          # Transparency for better visualization
            c=df['JobSat'],        # Color by Age
            cmap='viridis',     # Color map for bubble colors
            edgecolors='w',     # White edge color for bubbles
            linewidth=0.5)
plt.title('Age vs. Frequency of Participation with JobSat as Bubble Size', fontsize=14)
plt.xlabel(' Respondents’ age', fontsize=12)
plt.ylabel('Participation frequency', fontsize=12)

plt.colorbar(label='JobSat')
plt.grid(True)
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here
plt.figure(figsize=(8,6))
plt.scatter(df['ConvertedCompYearly'], df['JobSat'], 
            s=df['Age_numeric']*5,
            alpha=0.6,          # Transparency for better visualization
            c=df['Age_numeric'],        # Color by Age
            cmap='viridis',     # Color map for bubble colors
            edgecolors='w',     # White edge color for bubbles
            linewidth=0.5)
plt.title('Age vs. Frequency of Participation with JobSat as Bubble Size', fontsize=14)
plt.xlabel(' Respondents’ age', fontsize=12)
plt.ylabel('Participation frequency', fontsize=12)

plt.colorbar(label='JobSat')
plt.grid(True)
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here
grouped_df = df.groupby(['LanguageHaveWorkedWith', 'Age_numeric']).size().reset_index(name='Count')
top20_languages = grouped_df['LanguageHaveWorkedWith'].value_counts().head(20).index
grouped_df = grouped_df[grouped_df['LanguageHaveWorkedWith'].isin(top20_languages)]

plt.figure(figsize=(8,6))
plt.scatter(grouped_df['LanguageHaveWorkedWith'], grouped_df['Age_numeric'], 
            s=grouped_df['Count'],
            alpha=0.6,          # Transparency for better visualization
            c=grouped_df['Count'],        # Color by Age
            cmap='viridis',     # Color map for bubble colors
            edgecolors='w',     # White edge color for bubbles
            linewidth=0.5)
plt.title('Popularity of programming languages across age groups', fontsize=14)
plt.xlabel(' LanguageHaveWorkedWith', fontsize=12)
plt.ylabel('Age_numeric', fontsize=12)
plt.xticks(rotation=90)

plt.colorbar(label='freq')
plt.grid(True)
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
##Write your code here
grouped_df = df.groupby(['DatabaseWantToWorkWith', 'JobSat']).size().reset_index(name='Count')
top20_databases = grouped_df['DatabaseWantToWorkWith'].value_counts().head(20).index
grouped_df = grouped_df[grouped_df['DatabaseWantToWorkWith'].isin(top20_databases)]

plt.figure(figsize=(8,6))
plt.scatter(grouped_df['DatabaseWantToWorkWith'], grouped_df['JobSat'], 
            s=grouped_df['Count'],
            alpha=0.6,          # Transparency for better visualization
            c=grouped_df['Count'],        # Color by Age
            cmap='viridis',     # Color map for bubble colors
            edgecolors='w',     # White edge color for bubbles
            linewidth=0.5)
plt.title('Relationship between preferred databases and job satisfaction', fontsize=14)
plt.xlabel(' DatabaseWantToWorkWith', fontsize=12)
plt.ylabel('Job Satisfaction', fontsize=12)
plt.xticks(rotation=90)

plt.colorbar(label='freq')
plt.grid(True)
plt.show()
```
