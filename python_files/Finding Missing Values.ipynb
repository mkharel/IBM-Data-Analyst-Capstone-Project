### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
print(df.info())
print(df.describe())
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
df.isna().sum()
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
plt.figure(figsize=(18,12))
sns.heatmap(df.isna(), cmap='Blues', cbar=False, yticklabels=False)
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
print(df['Employment'].isna().sum())
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
df['RemoteWork'].value_counts().index[0]
#df['Employment'].mode()[0]
df['RemoteWork'].value_counts()
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
df['RemoteWork'] = df['RemoteWork'].fillna(df['RemoteWork'].mode()[0])
remoteWork = df['RemoteWork'].value_counts().reset_index()
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here     taken RemoteWork instead of Employment because Employment column has zero missing value
plt.figure(figsize=(8,6))
plt.bar(remoteWork['RemoteWork'], remoteWork['count'])
plt.title(' Distribution of a RemoteWork after imputation')
plt.xlabel('Types of remotework')
plt.ylabel('Frequency')
plt.xticks(rotation=15)
plt.show()
```
