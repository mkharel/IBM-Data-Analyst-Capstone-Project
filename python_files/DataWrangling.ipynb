### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
# Write your code here
df.shape
summary = pd.DataFrame({
    'Data Type': df.dtypes,
    'Non-Null Count': df.notnull().sum(),
    'Null Count': df.isnull().sum(),
    'Null Percentage': (df.isnull().sum() / len(df)) * 100
})

# Display the summary
print(summary)
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
df_encoded = pd.get_dummies(df, columns=['Employment'], prefix=['Employment'])
df_encoded
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
df.isnull().sum().sort_values(ascending=False).index[0]
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].median())
df['ConvertedCompYearly']
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
df['RemoteWork'].fillna(df['RemoteWork'].value_counts().index[0])
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
min_value = df['ConvertedCompYearly'].min()
max_value = df['ConvertedCompYearly'].max()
df['ConvertedCompYearly'] = (df['ConvertedCompYearly'] - min_value) / (max_value - min_value)
df['ConvertedCompYearly']
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
import numpy as np
# /opt/conda/lib/python3.12/site-packages/pandas/core/arraylike.py:399: RuntimeWarning: divide by zero encountered in log
#  result = getattr(ufunc, method)(*inputs, **kwargs) ===>>> log1p handles this problem
ConvertedCompYearly_log = np.log1p(df['ConvertedCompYearly'])
ConvertedCompYearly_log
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
#print(df['YearsCodePro'].value_counts())
df['YearsCodePro'] = df['YearsCodePro'].replace('Less than 1 year', 1)
df['YearsCodePro'] = df['YearsCodePro'].replace('More than 50 years', 50)
#df['YearsCodePro'].value_counts()

df['ExperienceLevel'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
df['ExperienceLevel'] = df['ExperienceLevel'].fillna(df['ExperienceLevel'].median())
df['ExperienceLevel'].value_counts()
```
