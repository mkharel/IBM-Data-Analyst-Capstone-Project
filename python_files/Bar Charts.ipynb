### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
## Write your code here

plt.figure(figsize=(8,6))
plt.hist(df['ConvertedCompYearly'])

plt.title('Distribution of yearly compensation')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('Frequency')

plt.yscale('log')
plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
## Write your code here
age_mapping = {
                "25-34 years old": 29.5,
                "35-44 years old": 39.5,
                "18-24 years old": 19.5,
                "45-54 years old": 49.5,
                "55-64 years old": 59.5,  
                "Under 18 years old": 17,
                "65 years or older": 65,
                "Prefer not to say": 0  
             }
df['Age_numeric'] = df['Age'].map(age_mapping)
```

### Task 2: Visualizing Relationships in Data


##### 1. Scatter Plot of `Age_numeric` and `ConvertedCompYearly`


Explore the relationship between age and compensation.



```python
## Write your code here
plt.figure(figsize=(8,6))
plt.scatter(df['ConvertedCompYearly'], df['Age_numeric'])

plt.title('Relationship between Age and Compensation')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('Age')

plt.grid(True)
plt.show()
```

##### 2. Bubble Plot of `ConvertedCompYearly` and `JobSatPoints_6` with `Age_numeric` as Bubble Size


Explore how compensation and job satisfaction are related, with age as the bubble size.


```python
## Write your code here
plt.figure(figsize=(8,6))
plt.scatter(df['ConvertedCompYearly'], df['JobSatPoints_6'],
           s=df['Age_numeric']*5,
            alpha=0.6,          # Transparency for better visualization
            c=df['Age_numeric'],        # Color by Age
            cmap='viridis',     # Color map for bubble colors
            edgecolors='w',     # White edge color for bubbles
            linewidth=0.5)

plt.title('Relationship between compensation and job satisfaction with age')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('JobSatPoints_6')

plt.colorbar(label='JobSat')
plt.grid(True)
plt.show()
```

### Task 3: Visualizing Composition of Data with Bar Charts


##### 1. Horizontal Bar Chart of `MainBranch` Distribution


Visualize the distribution of respondents’ primary roles to understand their professional focus.



```python
## Write your code here
mainbranch_dist = df['MainBranch'].value_counts()

plt.figure(figsize=(6,6))
plt.barh(mainbranch_dist.index, mainbranch_dist.values, color='skyblue')

# Set plot labels and title
plt.title('Distribution of Primary Roles (MainBranch)', fontsize=16)
plt.xlabel('Number of Respondents', fontsize=12)
plt.ylabel('Primary Role', fontsize=12)

# Display the plot
#plt.tight_layout()
plt.show()
```

##### 2. Vertical Bar Chart of Top 5 Programming Languages Respondents Want to Work With


Identify the most desired programming languages based on `LanguageWantToWorkWith`.



```python
## Write your code here
LanguageWantToWorkWith_dist = df['LanguageWantToWorkWith'].value_counts().head(25)

plt.figure(figsize=(8,6))
plt.bar(LanguageWantToWorkWith_dist.index, LanguageWantToWorkWith_dist.values, color='skyblue')

plt.title('Top 25 Programming Languages Respondents Want to Work With', fontsize=16)
plt.xlabel('Programming Languages', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

##### 3. Stacked Bar Chart of Median `JobSatPoints_6` and `JobSatPoints_7` by Age Group


Compare job satisfaction metrics across different age groups with a stacked bar chart.


```python
## Write your code here
grouped_df = df.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()

ax = grouped_df.plot(kind='bar', stacked=True, figsize=(10, 6), color=['lightblue', 'lightgreen'])

# Set plot labels and title
plt.title('Median Job Satisfaction Scores by Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Median Job Satisfaction Points', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

##### 4. Bar Chart of Database Popularity (`DatabaseHaveWorkedWith`)


Identify the most commonly used databases among respondents by visualizing `DatabaseHaveWorkedWith`.



```python
## Write your code here
Database_count = df['DatabaseHaveWorkedWith'].value_counts().head(25)

plt.figure(figsize=(8,6))
plt.bar(Database_count.index, Database_count.values, color='skyblue')

plt.title('Top 25 Database Respondents have worked With', fontsize=16)
plt.xlabel('Databases', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Comparison of Data with Bar Charts


##### 1. Grouped Bar Chart of Median `ConvertedCompYearly` for Different Age Groups


Compare median compensation across multiple age groups with a grouped bar chart.



```python
## Write your code here
age_grouped = df.groupby('Age')['ConvertedCompYearly'].median()

# Plotting the grouped bar chart
plt.figure(figsize=(10, 6))
age_grouped.plot(kind='bar', color='skyblue')

# Set plot labels and title
plt.title('Median Compensation by Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Median Compensation (Yearly)', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

##### 2. Bar Chart of Respondent Count by Country


Show the distribution of respondents by country to see which regions are most represented.



```python
## Write your code here
Country_count = df['Country'].value_counts()

plt.figure(figsize=(18,10))
plt.bar(Country_count.index, Country_count.values, color='skyblue')

plt.title('Distribution of respondents by country', fontsize=16)
plt.xlabel('Countries', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)

plt.xticks(rotation=90, fontsize=8)
plt.xlim(-2) 
plt.yscale('log')

plt.tight_layout()
plt.show()
```
